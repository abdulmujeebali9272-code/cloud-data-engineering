THE WITH STATEMENT


````python
Python mein **`with` statement** resources ko automatically manage karne ke liye use hota hai. Isse file, database connection, lock, etc. ka cleanup automatically ho jata hai.

### Without `with`

```python
file = open("data.txt", "r")
data = file.read()
file.close()   # manually close karna padta hai
```

Agar error aa jaye to `close()` shayad execute na ho.

### With `with`

```python
with open("data.txt", "r") as file:
    data = file.read()
    print(data)
```

Yahan:

* `open("data.txt", "r")` file open karta hai.
* `as file` file object ko variable `file` mein store karta hai.
* Block ke khatam hote hi file automatically close ho jati hai, chahe error aaye ya na aaye.

### Example

```python
with open("test.txt", "w") as f:
    f.write("Hello World")

print("File saved")
```

### Benefits

✅ Code cleaner hota hai.
✅ Resource leak nahi hoti.
✅ Automatic cleanup hota hai.
✅ Exception handling mein bhi safe rehta hai.

Agar aap beginner hain, to sabse common use **files ke saath `with open(...) as ...:`** hai. Main `with` statement ke peeche ka mechanism (`__enter__` aur `__exit__`) bhi samjha sakta hoon agar aap advanced detail chahte hain.
````

### Aasaan Zuban Mein

`self` **current object** ko refer karta hai.

Jab aap class se object banate ho, har object ki apni alag values hoti hain. `self` Python ko batata hai ke kis object ki baat ho rahi hai.

### Example

```python
class Student:
    def __init__(self, name):
        self.name = name

    def show_name(self):
        print(self.name)

s1 = Student("Ali")
s2 = Student("Ahmed")

s1.show_name()
s2.show_name()
```

Output:

```text
Ali
Ahmed
```

Yahan:

* `s1` ke liye `self` = `s1`
* `s2` ke liye `self` = `s2`

Jab:

```python
s1.show_name()
```

chalti hai, Python internally aise samajhta hai:

```python
Student.show_name(s1)
```

Aur jab:

```python
s2.show_name()
```

chalti hai, Python internally aise samajhta hai:

```python
Student.show_name(s2)
```

Isliye `self.name` har object ki apni value access karta hai.
