```python
def sum(a,b):
    c = a+b 
    return c

value = sum(4,3) # positional arguments 
print(value)

print(value * 2)
```

```python
def sum(a,b):
    c = a+b 
    d = a-b
    return c, d # return is an endpoint for a function

    print("Hello") # won't run

value1, value2 = sum(b=5,a=6) # Keyword arguments 
print(value1, value2)
```

```python
def apply_tax(amount, rate=0.15): # default parameter
    return amount + (amount * rate)

print(apply_tax(1000))          # uses default rate 0.15
```

```python
def apply_tax(amount, rate=0.15, discount=0.05): # default parameter
    return amount + (amount * rate) - (amount *discount )

print(apply_tax(1000))          
```

```python
print(apply_tax(1000, 0.10))    # overrides with 0.10
```
